```python
# ==============================================================================
# PREPROCESAMIENTO AVANZADO DE TEXTO (NLTK vs spaCy)
# ==============================================================================

import spacy
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Cargar modelo de spaCy (solo la primera vez puede tardar)
nlp = spacy.load("en_core_web_sm")

# ==============================================================================
# FUNCIÓN DE PREPROCESAMIENTO CON NLTK
# ==============================================================================
def preprocess_nltk(text):
    if not isinstance(text, str):
        return ""
    
    # Limpieza básica con regex
    text = text.lower()
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)
    text = re.sub(r'@\w+|#', '', text)
    text = re.sub(r'[^a-z\s]', '', text)
    text = text.strip()
    
    # Tokenización y eliminación de stopwords
    tokens = word_tokenize(text)
    
    stop_words = set(stopwords.words('english'))
    wine_stopwords = {'wine', 'aroma', 'flavor', 'note', 'finish', 'palate', 
                      'drink', 'taste', 'notes', 'flavors', 'aromas'}
    stop_words = stop_words.union(wine_stopwords)
    
    tokens = [word for word in tokens if word not in stop_words and len(word) > 2]
    
    # Lematización
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    
    return ' '.join(tokens)


# ==============================================================================
# FUNCIÓN DE PREPROCESAMIENTO CON spaCy
# ==============================================================================
def preprocess_spacy(text):
    if not isinstance(text, str):
        return ""
    
    # Limpieza básica con regex
    text = text.lower()
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)
    text = re.sub(r'@\w+|#', '', text)
    text = re.sub(r'[^a-z\s]', '', text)
    text = text.strip()
    
    # Procesamiento con spaCy
    doc = nlp(text)
    
    tokens = [
        token.lemma_ for token in doc 
        if not token.is_stop 
        and not token.is_punct 
        and not token.is_digit
        and len(token.lemma_) > 2
    ]
    
    wine_stop = {'wine', 'aroma', 'flavor', 'note', 'finish', 'palate', 
                 'drink', 'taste', 'notes', 'flavors', 'aromas'}
    tokens = [t for t in tokens if t not in wine_stop]
    
    return ' '.join(tokens)


# ==============================================================================
# APLICACIÓN DEL PREPROCESAMIENTO
# ==============================================================================
print("🚀 Aplicando preprocesamiento avanzado (NLTK y spaCy)...\n")

df_nlp['description_nltk'] = df_nlp['description'].apply(preprocess_nltk)
df_nlp['description_spacy'] = df_nlp['description'].apply(preprocess_spacy)

# Comparación visual de resultados
print("=== EJEMPLO COMPARATIVO DE PREPROCESAMIENTO ===\n")
idx = 0  # Puedes cambiar este índice para ver otros ejemplos

print("Texto Original:")
print(df_nlp['description'].iloc[idx][:400] + "...\n")

print("Después de NLTK:")
print(df_nlp['description_nltk'].iloc[idx][:300] + "...\n")

print("Después de spaCy:")
print(df_nlp['description_spacy'].iloc[idx][:300] + "...")

print("\n✅ Preprocesamiento completado. Se crearon las columnas 'description_nltk' y 'description_spacy'.")
```

    <p style="color: #f2f2f2; font-size: 14px; line-height: 1.7;">
        <strong style="color: #D99B77;">¿Por qué <code>max_features=8000</code>?</strong><br>
        Para captura más matices semánticos pero incrementa la dimensionalidad y el costo computacional.
        Con 8.000 features se logra un equilibrio entre expresividad y eficiencia para un corpus de ~100k reseñas.

        <strong style="color: #D99B77;">¿Por qué <code>min_df=5</code> y <code>max_df=0.85</code>?</strong><br>
        <code>min_df=5</code> elimina términos excesivamente raros (errores tipográficos, nombres propios únicos)
        que no generalizan bien. <code>max_df=0.85</code> descarta palabras presentes en más del 85% de los documentos,
        complementando la penalización natural del IDF sobre términos muy frecuentes.

<div style="background-color: #262524; padding: 25px; border-radius: 12px; border-left: 10px solid #F23838; font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif; box-shadow: 2px 2px 15px rgba(0,0,0,0.3); max-width: 100%; box-sizing: border-box;">

        <span style="font-size: 30px; margin-right: 15px;">🧠</span>
        <h1 style="color: #D99B77; margin: 0; font-size: 22px; text-transform: uppercase; letter-spacing: 2px;">
            Arquitectura del Modelo: TextClassifier
        </h1>
    </div>
    <hr style="border: 0; border-top: 1px solid #8C533E; margin-bottom: 15px;">
    <p style="color: #f2f2f2; font-size: 15px; line-height: 1.7; margin: 0;">
        Con los vectores TF-IDF listos, se construye la red neuronal profunda (<strong>DNN</strong>) en
        <strong>PyTorch</strong> que clasificará las reseñas en una de las 10 variedades de vino.
        La arquitectura es de tipo <em>fully-connected</em> (todas las neuronas de una capa conectadas con la siguiente).
    </p>
    <h3 style="color: #D99B77; margin: 18px 0 10px 0;">Componentes de la arquitectura:</h3>
    <ul style="color: #f2f2f2; font-size: 15px; line-height: 1.8;">
        <li><strong>Capas densas (<code>nn.Linear</code>):</strong> tres capas lineales con dimensiones
            decrecientes (input → 512 → 256 → num_classes) para ir comprimiendo la representación de forma progresiva.</li>
        <li><strong>Batch Normalization (<code>nn.BatchNorm1d</code>):</strong> normaliza las activaciones
            de cada mini-batch, acelerando la convergencia y reduciendo la sensibilidad a la inicialización de pesos.</li>
        <li><strong>Dropout (<code>p=0.5</code>):</strong> desactiva aleatoriamente el 50% de las neuronas
            durante el entrenamiento, forzando al modelo a aprender representaciones más robustas y reduciendo el sobreajuste.</li>
        <li><strong>ReLU:</strong> función de activación no lineal que permite al modelo aprender
            relaciones complejas entre los features del texto.</li>
    </ul>
    <div style="background-color: rgba(217, 155, 119, 0.1); padding: 10px; border-radius: 8px; border: 1px dashed #D99B77; margin-top: 18px;">
        <p style="color: #D99B77; margin: 0; font-size: 14px;">
            <strong>Nota:</strong> la capa de salida (<code>fc3</code>) no aplica <code>Softmax</code> explícitamente,
            ya que <code>nn.CrossEntropyLoss</code> lo incorpora internamente de forma numéricamente estable.
        </p>
    </div>
</div>

```python
# ==============================================================================
# ARQUITECTURA DEL MODELO: TextClassifier
# ==============================================================================

class TextClassifier(nn.Module):
    """
    Red neuronal fully-connected para clasificación multiclase de notas de cata.

    Arquitectura:
        input_dim → 512 → 256 → num_classes

    Técnicas de regularización:
        - Batch Normalization: estabiliza y acelera el entrenamiento.
        - Dropout (p=0.5): reduce el sobreajuste desactivando neuronas aleatoriamente.
        - ReLU: introduce no-linealidad entre capas.
    """
    def __init__(self, input_dim, hidden_dim1=512, hidden_dim2=256, num_classes=10, dropout_p=0.5):
        super(TextClassifier, self).__init__()

        # --- Capa oculta 1 ---
        self.fc1      = nn.Linear(input_dim, hidden_dim1)
        self.bn1      = nn.BatchNorm1d(hidden_dim1)
        self.dropout1 = nn.Dropout(p=dropout_p)

        # --- Capa oculta 2 ---
        self.fc2      = nn.Linear(hidden_dim1, hidden_dim2)
        self.bn2      = nn.BatchNorm1d(hidden_dim2)
        self.dropout2 = nn.Dropout(p=dropout_p)

        # --- Capa de salida (sin Softmax: lo aplica CrossEntropyLoss internamente) ---
        self.fc3  = nn.Linear(hidden_dim2, num_classes)

        self.relu = nn.ReLU()

    def forward(self, x):
        # Capa 1: Linear → BatchNorm → ReLU → Dropout
        x = self.dropout1(self.relu(self.bn1(self.fc1(x))))

        # Capa 2: Linear → BatchNorm → ReLU → Dropout
        x = self.dropout2(self.relu(self.bn2(self.fc2(x))))

        # Capa de salida: logits
        return self.fc3(x)


# ------------------------------------------------------------------------------
# Instanciación del modelo
# ------------------------------------------------------------------------------
input_dim   = X_train.shape[1]       # features TF-IDF (definido en la sección anterior)
num_classes = len(le.classes_)        # variedades únicas

model = TextClassifier(
    input_dim=input_dim,
    hidden_dim1=512,
    hidden_dim2=256,
    num_classes=num_classes,
    dropout_p=0.5
).to(device)

# ------------------------------------------------------------------------------
# Resumen del modelo
# ------------------------------------------------------------------------------
total_params     = sum(p.numel() for p in model.parameters())
trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)

print("✅ Modelo TextClassifier instanciado correctamente")
print(f"   Input dim        : {input_dim:,}")
print(f"   Clases           : {num_classes}  → {list(le.classes_)}")
print(f"   Parámetros totales    : {total_params:,}")
print(f"   Parámetros entrenables: {trainable_params:,}")
print(f"   Dispositivo      : {device}")
print(f"\n{model}")
```

    <h3 style="color: #D99B77; margin-top: 0;">💡 Reflexión: decisiones de diseño en la arquitectura</h3>
    <p style="color: #f2f2f2; font-size: 14px; line-height: 1.7;">
        <strong style="color: #D99B77;">¿Por qué dimensiones decrecientes (512 → 256)?</strong><br>
        Reducir progresivamente el tamaño de las capas ocultas obliga al modelo a comprimir la información
        en representaciones cada vez más abstractas y compactas, similar al funcionamiento de un autoencoder.
        Esto favorece la generalización y evita memorizar el conjunto de entrenamiento.
    </p>
    <p style="color: #f2f2f2; font-size: 14px; line-height: 1.7; margin-top: 10px;">
        <strong style="color: #D99B77;">Batch Normalization antes o después de ReLU:</strong><br>
        Se aplica BatchNorm antes de la activación (orden: Linear → BN → ReLU → Dropout). Este orden
        normaliza las pre-activaciones y permite que ReLU opere sobre una distribución más estable.
    </p>
    <p style="color: #f2f2f2; font-size: 14px; line-height: 1.7; margin-top: 10px;">
        <strong style="color: #D99B77;">Limitación principal:</strong><br>
        Al trabajar con vectores TF-IDF estáticos, la red no tiene acceso al orden ni al contexto
        semántico de las palabras. Una arquitectura basada en Transformers (DistilBERT)
        podría capturar esas relaciones y mejorar el rendimiento significativamente.
    </p>

    </p>
    <h3 style="color: #D99B77; margin: 18px 0 10px 0;">Configuración del entrenamiento:</h3>
    <ul style="color: #f2f2f2; font-size: 15px; line-height: 1.8;">
        <li><strong>Optimizador Adam:</strong> algoritmo adaptativo que ajusta la tasa de aprendizaje
            individualmente para cada parámetro, combinando las ventajas de RMSProp y SGD con momentum.</li>
        <li><strong>CrossEntropyLoss:</strong> función de pérdida estándar para clasificación multiclase,
            que incorpora Softmax internamente y es numéricamente más estable que aplicarlo por separado.</li>
        <li><strong>Early Stopping (paciencia = 5):</strong> si la pérdida de validación no mejora
            durante 5 épocas consecutivas, el entrenamiento se detiene automáticamente y se restaura
            el mejor modelo guardado, evitando el sobreajuste.</li>
        <li><strong>Scheduler ReduceLROnPlateau:</strong> reduce la tasa de aprendizaje a la mitad
            cuando la pérdida de validación se estanca, permitiendo ajustes más finos en las etapas finales.</li>
    </ul>
    <div style="background-color: rgba(217, 155, 119, 0.1); padding: 10px; border-radius: 8px; border: 1px dashed #D99B77; margin-top: 18px;">
        <p style="color: #D99B77; margin: 0; font-size: 14px;">
            <strong>Nota:</strong> el mejor modelo se guarda en disco (<code>best_model.pth</code>) cada vez que
            la pérdida de validación mejora. Al finalizar el entrenamiento, ese checkpoint se recarga
            automáticamente para garantizar que la evaluación final use los mejores pesos encontrados.
        </p>
    </div>
</div>

```python
# ==============================================================================
# ENTRENAMIENTO CON EARLY STOPPING + SCHEDULER
# ==============================================================================

# ------------------------------------------------------------------------------
# Hiperparámetros
# ------------------------------------------------------------------------------
num_epochs    = 50
patience      = 5        # épocas sin mejora antes de detener el entrenamiento
learning_rate = 1e-3

# ------------------------------------------------------------------------------
# Optimizador, función de pérdida y scheduler
# ------------------------------------------------------------------------------
optimizer = optim.Adam(model.parameters(), lr=learning_rate)
criterion = nn.CrossEntropyLoss()

# Reduce lr x0.5 si val_loss no mejora en 3 épocas consecutivas
scheduler = optim.lr_scheduler.ReduceLROnPlateau(
    optimizer, mode='min', factor=0.5, patience=3, verbose=True
)

# ------------------------------------------------------------------------------
# Variables de control
# ------------------------------------------------------------------------------
best_val_loss     = float('inf')
epochs_no_improve = 0
early_stop        = False
history           = {'train_loss': [], 'val_loss': [], 'train_acc': [], 'val_acc': []}

# ------------------------------------------------------------------------------
# Loop de entrenamiento
# ------------------------------------------------------------------------------
print("🚀 Iniciando entrenamiento...\n")
print(f"{'Época':>6} | {'Train Loss':>10} | {'Val Loss':>10} | {'Train Acc':>10} | {'Val Acc':>9} | {'LR':>8}")
print("-" * 70)

for epoch in range(num_epochs):

    # ==========================================================================
    # FASE DE ENTRENAMIENTO
    # ==========================================================================
    model.train()
    train_loss, train_correct, train_total = 0.0, 0, 0

    for inputs, labels in train_loader:
        optimizer.zero_grad()
        outputs = model(inputs)
        loss    = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        train_loss    += loss.item() * inputs.size(0)
        preds          = outputs.argmax(dim=1)
        train_correct += (preds == labels).sum().item()
        train_total   += inputs.size(0)

    train_loss /= train_total
    train_acc   = train_correct / train_total

    # ==========================================================================
    # FASE DE VALIDACIÓN
    # ==========================================================================
    model.eval()
    val_loss, val_correct, val_total = 0.0, 0, 0

    with torch.no_grad():
        for inputs, labels in val_loader:
            outputs    = model(inputs)
            loss       = criterion(outputs, labels)
            val_loss  += loss.item() * inputs.size(0)
            preds      = outputs.argmax(dim=1)
            val_correct += (preds == labels).sum().item()
            val_total  += inputs.size(0)

    val_loss /= val_total
    val_acc   = val_correct / val_total

    # Registrar historial
    history['train_loss'].append(train_loss)
    history['val_loss'].append(val_loss)
    history['train_acc'].append(train_acc)
    history['val_acc'].append(val_acc)

    # Scheduler step
    scheduler.step(val_loss)
    current_lr = optimizer.param_groups[0]['lr']

    # ==========================================================================
    # EARLY STOPPING
    # ==========================================================================
    if val_loss < best_val_loss:
        best_val_loss     = val_loss
        epochs_no_improve = 0
        torch.save(model.state_dict(), 'best_model.pth')
    else:
        epochs_no_improve += 1
        if epochs_no_improve >= patience:
            print(f"\n⏹️  Early Stopping activado en época {epoch + 1} "
                  f"(sin mejora por {patience} épocas consecutivas)")
            early_stop = True
            break

    # Mostrar progreso cada 5 épocas o en la primera
    if (epoch + 1) % 5 == 0 or epoch == 0:
        print(f"{epoch+1:>6} | {train_loss:>10.4f} | {val_loss:>10.4f} | "
              f"{train_acc:>9.2%} | {val_acc:>8.2%} | {current_lr:>8.2e}")

# ------------------------------------------------------------------------------
# Restaurar el mejor modelo
# ------------------------------------------------------------------------------
model.load_state_dict(torch.load('best_model.pth'))

if not early_stop:
    print(f"\n✅ Entrenamiento completado ({num_epochs} épocas sin activar Early Stopping)")
print(f"✅ Mejor modelo restaurado  (val_loss = {best_val_loss:.4f})")
```

    <h3 style="color: #D99B77; margin-top: 0;">💡 Reflexión: decisiones de diseño en el entrenamiento</h3>
    <p style="color: #f2f2f2; font-size: 14px; line-height: 1.7;">
        <strong style="color: #D99B77;">¿Por qué Adam y no SGD?</strong><br>
        Adam adapta la tasa de aprendizaje por parámetro en función del historial de gradientes,
        lo que lo hace especialmente efectivo en espacios de alta dimensionalidad como los vectores TF-IDF.
        SGD puede converger a mejores mínimos con una búsqueda exhaustiva de hiperparámetros,
        pero requiere más épocas y ajuste manual del learning rate.
    </p>
    <p style="color: #f2f2f2; font-size: 14px; line-height: 1.7; margin-top: 10px;">
        <strong style="color: #D99B77;">Early Stopping vs. número fijo de épocas:</strong><br>
        Fijar un número de épocas arbitrario puede llevar a sobreajuste si es demasiado alto
        o a underfitting si es demasiado bajo. Early Stopping delega esa decisión al comportamiento
        real de la pérdida de validación, deteniendo el entrenamiento en el punto óptimo
        y restaurando automáticamente los mejores pesos encontrados.
    </p>
    <p style="color: #f2f2f2; font-size: 14px; line-height: 1.7; margin-top: 10px;">
        <strong style="color: #D99B77;">ReduceLROnPlateau como complemento:</strong><br>
        Cuando el modelo se acerca a un mínimo, gradientes grandes pueden hacer que "salte" por encima.
        Reducir el learning rate en esos momentos permite ajustes más finos sin necesidad de
        reiniciar el entrenamiento, mejorando la convergencia final antes de que Early Stopping se active.
    </p>
</div>

```python
# ==============================================================================
# EVALUACIÓN FINAL SOBRE CONJUNTO DE TEST
# ==============================================================================

model.eval()
y_pred, y_true = [], []

with torch.no_grad():
    for inputs, labels in test_loader:
        outputs = model(inputs)
        preds   = outputs.argmax(dim=1)
        y_pred.extend(preds.cpu().numpy())
        y_true.extend(labels.cpu().numpy())

y_pred = np.array(y_pred)
y_true = np.array(y_true)

# Accuracy global
test_acc = (y_pred == y_true).mean()
print(f"✅ Accuracy en Test: {test_acc:.4f}  ({test_acc:.2%})\n")

# ------------------------------------------------------------------------------
# Matriz de confusión
# ------------------------------------------------------------------------------
cm = confusion_matrix(y_true, y_pred)

fig, ax = plt.subplots(figsize=(12, 9))
fig.patch.set_facecolor('#262524')
ax.set_facecolor('#1a1918')

sns.heatmap(
    cm,
    annot=True,
    fmt='d',
    cmap='YlOrRd',
    xticklabels=le.classes_,
    yticklabels=le.classes_,
    linewidths=0.5,
    linecolor='#262524',
    ax=ax
)

ax.set_title('Matriz de Confusión — Conjunto de Test', color='#D99B77', fontsize=14, pad=15)
ax.set_xlabel('Predicción',  color='#f2f2f2', fontsize=12)
ax.set_ylabel('Valor Real',  color='#f2f2f2', fontsize=12)
ax.tick_params(colors='#f2f2f2', rotation=45)

plt.tight_layout()
plt.show()
```
