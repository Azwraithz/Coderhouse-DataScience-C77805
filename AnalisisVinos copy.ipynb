```python
import re
import nltk
import spacy
import pandas as pd
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Descargas necesarias para NLTK
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')

# Cargamos el modelo de spaCy (Asegúrate de haberlo instalado con !python -m spacy download en_core_web_sm)
try:
    nlp = spacy.load('en_core_web_sm')
except:
    import os
    os.system('python -m spacy download en_core_web_sm')
    nlp = spacy.load('en_core_web_sm')

# 1. Función de Limpieza con Regex
def clean_text(text):
    # Convertir a minúsculas
    text = text.lower()
    # Eliminar URLs
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    # Eliminar menciones y hashtags (aunque no abundan en este dataset, cumple la consigna)
    text = re.sub(r'\@\w+|\#','', text)
    # Eliminar caracteres especiales y números (solo dejamos letras y espacios)
    text = re.sub(r'[^a-z\s]', '', text)
    return text.strip()

# 2. Lematización Comparativa   
def comparative_lemmatization(text):
    # --- Versión NLTK ---
    lemmatizer_nltk = WordNetLemmatizer()
    stop_words = set(stopwords.words('english'))
    tokens_nltk = nltk.word_tokenize(text)
    # Lematizamos solo si no es una stopword
    lems_nltk = [lemmatizer_nltk.lemmatize(w) for w in tokens_nltk if w not in stop_words]
    
    # --- Versión spaCy ---
    doc = nlp(text)
    # spaCy detecta automáticamente stop_words y lematiza con base en el contexto (POS tagging)
    lems_spacy = [token.lemma_ for token in doc if not token.is_stop and not token.is_punct]
    
    return " ".join(lems_nltk), " ".join(lems_spacy)

# Aplicamos la limpieza inicial
df_nlp['description_clean'] = df_nlp['description'].apply(clean_text)

# Tomamos un ejemplo para la comparativa que pide la consigna
ejemplo = df_nlp['description_clean'].iloc[0]
nltk_res, spacy_res = comparative_lemmatization(ejemplo)

print("--- COMPARATIVA DE LEMATIZACIÓN ---")
print(f"Texto Original: {ejemplo[:100]}...")
print(f"\nResultado NLTK:\n{nltk_res[:100]}...")
print(f"\nResultado spaCy:\n{spacy_res[:100]}...")

# Para el procesamiento del dataset completo, usaremos spaCy por su precisión contextual
# Nota: Esto puede tardar un par de minutos dependiendo del tamaño del dataset
def final_process(text):
    doc = nlp(text)
    return " ".join([token.lemma_ for token in doc if not token.is_stop])

print("\nProcesando el dataset completo con spaCy... por favor espera.")
df_nlp['processed_text'] = df_nlp['description_clean'].apply(final_process)

# Visualización final de la transformación
display(df_nlp[['description', 'processed_text']].head().style.set_properties(**{
    'text-align': 'left',
    'background-color': '#262524',
    'color': '#D99B77',
    'border': '1px solid #8C533E'
}))
```

```python
def preprocess_text(text):
    """Preprocesamiento completo para notas de cata de vinos."""
    if not isinstance(text, str):
        return ""
    
    # 1. Limpieza básica
    text = text.lower()
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)
    text = re.sub(r'\@\w+|\#', '', text)
    text = re.sub(r'[^a-z\s]', '', text)
    text = text.strip()
    
    # 2. Tokenización
    tokens = word_tokenize(text)
    
    # 3. Stopwords (inglés + específicas de vino)
    stop_words = set(stopwords.words('english'))
    wine_stopwords = {'wine', 'aroma', 'flavor', 'note', 'finish', 'palate', 'drink', 'taste', 'notes'}
    stop_words = stop_words.union(wine_stopwords)
    
    tokens = [word for word in tokens if word not in stop_words and len(word) > 2]
    
    # 4. Lematización
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    
    return ' '.join(tokens)

# Aplicación
print("🚀 Aplicando preprocesamiento avanzado...")
df_nlp['description_clean'] = df_nlp['description'].apply(preprocess_text)

# Validación visual
print("\nEjemplo antes y después:")
print("Original →", df_nlp['description'].iloc[0][:200] + "...")
print("Procesado →", df_nlp['description_clean'].iloc[0][:200] + "...")
```

```python
import spacy
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Cargar modelo de spaCy (ejecutar una sola vez si es necesario)
# !python -m spacy download en_core_web_sm
nlp = spacy.load("en_core_web_sm")

# ==============================================================================
# FUNCIÓN DE PREPROCESAMIENTO CON NLTK
# ==============================================================================
def preprocess_nltk(text):
    """Preprocesamiento usando NLTK (tokenización + lematización)."""
    if not isinstance(text, str):
        return ""
    
    # 1. Limpieza básica con regex
    text = text.lower()
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)   # URLs
    text = re.sub(r'\@\w+|\#', '', text)                  # menciones y hashtags
    text = re.sub(r'[^a-z\s]', '', text)                  # solo letras y espacios
    text = text.strip()
    
    # 2. Tokenización
    tokens = word_tokenize(text)
    
    # 3. Eliminar stopwords + palabras cortas + términos poco informativos de vino
    stop_words = set(stopwords.words('english'))
    wine_stopwords = {'wine', 'aroma', 'flavor', 'note', 'finish', 'palate', 'drink', 'taste', 'notes'}
    stop_words = stop_words.union(wine_stopwords)
    
    tokens = [word for word in tokens if word not in stop_words and len(word) > 2]
    
    # 4. Lematización con NLTK
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    
    return ' '.join(tokens)


# ==============================================================================
# FUNCIÓN DE PREPROCESAMIENTO CON spaCy
# ==============================================================================
def preprocess_spacy(text):
    """Preprocesamiento usando spaCy (más preciso en contexto)."""
    if not isinstance(text, str):
        return ""
    
    # 1. Limpieza básica
    text = text.lower()
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)
    text = re.sub(r'\@\w+|\#', '', text)
    text = re.sub(r'[^a-z\s]', '', text)
    text = text.strip()
    
    # 2. Procesamiento con spaCy
    doc = nlp(text)
    
    # 3. Lematización + filtrado
    tokens = [
        token.lemma_ for token in doc 
        if not token.is_stop 
        and not token.is_punct 
        and not token.is_digit
        and len(token.lemma_) > 2
    ]
    
    # Stopwords específicas del dominio
    wine_stop = {'wine', 'aroma', 'flavor', 'note', 'finish', 'palate', 'drink', 'taste', 'notes'}
    tokens = [t for t in tokens if t not in wine_stop]
    
    return ' '.join(tokens)


# ==============================================================================
# APLICACIÓN DEL PREPROCESAMIENTO
# ==============================================================================
print("🚀 Aplicando preprocesamiento avanzado (NLTK y spaCy)...")

df_nlp['description_nltk'] = df_nlp['description'].apply(preprocess_nltk)
df_nlp['description_spacy'] = df_nlp['description'].apply(preprocess_spacy)

# Mostrar comparación visual
print("\n=== EJEMPLO COMPARATIVO DE PREPROCESAMIENTO ===")
idx = 42  # puedes cambiar este índice
print("Texto Original:")
print(df_nlp['description'].iloc[idx][:300] + "...\n")

print("Después de NLTK:")
print(df_nlp['description_nltk'].iloc[idx][:250] + "...\n")

print("Después de spaCy:")
print(df_nlp['description_spacy'].iloc[idx][:250] + "...")
```

        Implementamos una red neuronal fully-connected en PyTorch siguiendo las consignas de la evaluación final:
    </p>
    <ul style="color: #f2f2f2; font-size: 14px;">
        <li>Al menos dos capas densas (<code>nn.Linear</code>)</li>
        <li>Batch Normalization para estabilizar y acelerar el entrenamiento</li>
        <li>Dropout con <strong>p = 0.5</strong> para reducir el sobreajuste</li>
        <li>ReLU como función de activación</li>
    </ul>
    <p style="color: #f2f2f2; font-size: 15px; margin-top: 10px;">
        La entrada corresponde a los vectores TF-IDF generados en la fase anterior.

```python
# ==============================================================================
# ARQUITECTURA DEL MODELO: TextClassifier
# ==============================================================================

class TextClassifier(nn.Module):
    """
    Red neuronal fully-connected para clasificación multiclase de notas de cata.
    Requisitos de la evaluación:
      - Al menos 2 capas densas
      - Batch Normalization
      - Dropout p=0.5
    """
    def __init__(self, input_dim, hidden_dim1=512, hidden_dim2=256, num_classes=10, dropout_p=0.5):
        super(TextClassifier, self).__init__()
        
        self.fc1 = nn.Linear(input_dim, hidden_dim1)
        self.bn1 = nn.BatchNorm1d(hidden_dim1)
        self.dropout1 = nn.Dropout(p=dropout_p)
        
        self.fc2 = nn.Linear(hidden_dim1, hidden_dim2)
        self.bn2 = nn.BatchNorm1d(hidden_dim2)
        self.dropout2 = nn.Dropout(p=dropout_p)
        
        self.fc3 = nn.Linear(hidden_dim2, num_classes)
        
        self.relu = nn.ReLU()
        
    def forward(self, x):
        # Capa 1
        x = self.fc1(x)
        x = self.bn1(x)
        x = self.relu(x)
        x = self.dropout1(x)
        
        # Capa 2
        x = self.fc2(x)
        x = self.bn2(x)
        x = self.relu(x)
        x = self.dropout2(x)
        
        # Capa de salida (sin softmax, se aplica en CrossEntropyLoss)
        x = self.fc3(x)
        return x


# ==============================================================================
# INSTANCIACIÓN DEL MODELO
# ==============================================================================

# Dimensiones
input_dim = X_train.shape[1]          # número de features del TF-IDF
num_classes = len(le.classes_)        # número de variedades

model = TextClassifier(
    input_dim=input_dim,
    hidden_dim1=512,
    hidden_dim2=256,
    num_classes=num_classes,
    dropout_p=0.5
)

# Mover a GPU si está disponible
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)

print(f"✅ Modelo TextClassifier creado correctamente")
print(f"   Dimensiones de entrada: {input_dim}")
print(f"   Número de clases: {num_classes}")
print(f"   Dispositivo: {device}")
```

        y Dropout (p=0.5) para mitigar el sobreajuste, tal como solicitaba la consigna. 
        Esta configuración ofrece un buen equilibrio entre capacidad de representación y regularización.

    <ul style="color: #f2f2f2;">

        <li>Función de pérdida <strong>CrossEntropyLoss</strong></li>
        <li><strong>Early Stopping</strong> con paciencia de 5 épocas (si la pérdida de validación no mejora)</li>

```python
# ==============================================================================
# ENTRENAMIENTO CON EARLY STOPPING
# ==============================================================================

# Hiperparámetros
num_epochs = 50
patience = 5                    # Early Stopping: cuántas épocas sin mejora
learning_rate = 0.001
batch_size = 64

# Convertir matrices sparse a dense (ya corregido)
X_train_dense = X_train.toarray() if hasattr(X_train, 'toarray') else X_train
X_val_dense   = X_val.toarray()   if hasattr(X_val,   'toarray') else X_val

X_train_t = torch.tensor(X_train_dense, dtype=torch.float32).to(device)
y_train_t = torch.tensor(y_train, dtype=torch.long).to(device)

X_val_t   = torch.tensor(X_val_dense,   dtype=torch.float32).to(device)
y_val_t   = torch.tensor(y_val,   dtype=torch.long).to(device)

# DataLoaders
train_dataset = TensorDataset(X_train_t, y_train_t)
val_dataset   = TensorDataset(X_val_t, y_val_t)

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader   = DataLoader(val_dataset,   batch_size=batch_size, shuffle=False)

# Optimizador y función de pérdida
optimizer = optim.Adam(model.parameters(), lr=learning_rate)
criterion = nn.CrossEntropyLoss()

# Early Stopping
best_val_loss = float('inf')
epochs_no_improve = 0
early_stop = False

history = {'train_loss': [], 'val_loss': []}

print("🚀 Iniciando entrenamiento con Early Stopping...\n")

for epoch in range(num_epochs):
    # ==================== TRAINING ====================
    model.train()
    train_loss = 0.0
    
    for inputs, labels in train_loader:
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        train_loss += loss.item() * inputs.size(0)
    
    train_loss /= len(train_loader.dataset)
    history['train_loss'].append(train_loss)
    
    # ==================== VALIDATION ====================
    model.eval()
    val_loss = 0.0
    
    with torch.no_grad():
        for inputs, labels in val_loader:
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            val_loss += loss.item() * inputs.size(0)
    
    val_loss /= len(val_loader.dataset)
    history['val_loss'].append(val_loss)
    
    # ==================== EARLY STOPPING ====================
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        epochs_no_improve = 0
        # Guardar el mejor modelo
        torch.save(model.state_dict(), 'best_model.pth')
    else:
        epochs_no_improve += 1
        if epochs_no_improve >= patience:
            print(f"⏹️ Early Stopping activado en la época {epoch+1}")
            early_stop = True
            break
    
    # Mostrar progreso cada 5 épocas
    if (epoch + 1) % 5 == 0 or epoch == 0:
        print(f"Época {epoch+1:2d}/{num_epochs} | "
              f"Train Loss: {train_loss:.4f} | "
              f"Val Loss: {val_loss:.4f} | "
              f"Mejor Val Loss: {best_val_loss:.4f}")

if not early_stop:
    print("✅ Entrenamiento completado sin activar Early Stopping")

# Cargar el mejor modelo guardado
model.load_state_dict(torch.load('best_model.pth'))
print("✅ Mejor modelo cargado desde 'best_model.pth'")
```

```python
# ==============================================================================
# 1. PREPARACIÓN DE TENSORS PARA PYTORCH
# ==============================================================================
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset

# Convertir matrices sparse a formato denso (necesario para PyTorch)
# .toarray() convierte csr_matrix → numpy dense array
X_train_dense = X_train.toarray() if hasattr(X_train, 'toarray') else X_train
X_val_dense   = X_val.toarray()   if hasattr(X_val,   'toarray') else X_val
X_test_dense  = X_test.toarray()  if hasattr(X_test,  'toarray') else X_test

# Convertir a tensores de PyTorch
X_train_t = torch.tensor(X_train_dense, dtype=torch.float32)
y_train_t = torch.tensor(y_train, dtype=torch.long)

X_val_t   = torch.tensor(X_val_dense,   dtype=torch.float32)
y_val_t   = torch.tensor(y_val,   dtype=torch.long)

X_test_t  = torch.tensor(X_test_dense,  dtype=torch.float32)
y_test_t  = torch.tensor(y_test,  dtype=torch.long)

print(f"✅ Tensores preparados correctamente")
print(f"   X_train: {X_train_t.shape} | y_train: {y_train_t.shape}")
print(f"   X_val:   {X_val_t.shape}   | y_val:   {y_val_t.shape}")
print(f"   X_test:  {X_test_t.shape}  | y_test:  {y_test_t.shape}")
```

```python
# ==============================================================================
# 2. CREACIÓN DE DATASETS Y DATALOADERS
# ==============================================================================
train_dataset = TensorDataset(X_train_t, y_train_t)
val_dataset   = TensorDataset(X_val_t,   y_val_t)
test_dataset  = TensorDataset(X_test_t,  y_test_t)

batch_size = 64

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader   = DataLoader(val_dataset,   batch_size=batch_size, shuffle=False)
test_loader  = DataLoader(test_dataset,  batch_size=batch_size, shuffle=False)

print(f"✅ DataLoaders creados con batch_size = {batch_size}")
```

```python
class TextClassifier(nn.Module):
    def __init__(self, input_dim, num_classes):
        super(TextClassifier, self).__init__()
        
        # Capa de entrada a 512 neuronas
        self.fc1 = nn.Linear(input_dim, 512)
        self.bn1 = nn.BatchNorm1d(512) # Normaliza las activaciones
        
        # Capa oculta a 128 neuronas
        self.fc2 = nn.Linear(512, 128)
        self.bn2 = nn.BatchNorm1d(128)
        
        # Capa de salida (una neurona por cada variedad de vino)
        self.fc3 = nn.Linear(128, num_classes)
        
        # Regularización y Activación
        self.dropout = nn.Dropout(p=0.5) # Dropout al 50% según consigna
        self.relu = nn.ReLU()
        
    def forward(self, x):
        # Bloque 1
        x = self.relu(self.bn1(self.fc1(x)))
        x = self.dropout(x)
        
        # Bloque 2
        x = self.relu(self.bn2(self.fc2(x)))
        
        # Salida
        x = self.fc3(x)
        return x

# Instanciamos el modelo
# input_dim es 2500 (características de TF-IDF)
# num_classes es la cantidad de variedades filtradas
model = TextClassifier(input_dim=X_train.shape[1], num_classes=len(le.classes_))
print(model)
```

```python
# Configuración del optimizador y la función de pérdida
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Parámetros de Early Stopping
patience = 5
best_val_loss = float('inf')
epochs_no_improve = 0
history = {'train_loss': [], 'val_loss': []}

epochs = 5

print("--- Iniciando Entrenamiento ---")
for epoch in range(epochs):
    # FASE DE ENTRENAMIENTO
    model.train()
    train_loss = 0
    for inputs, labels in train_loader:
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        train_loss += loss.item()
    
    train_loss /= len(train_loader)
    
    # FASE DE VALIDACIÓN
    model.eval()
    val_loss = 0
    with torch.no_grad():
        for inputs, labels in val_loader:
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            val_loss += loss.item()
    
    val_loss /= len(val_loader)
    
    history['train_loss'].append(train_loss)
    history['val_loss'].append(val_loss)
    
    print(f"Época {epoch+1:02d} | Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f}")
    
    # Lógica de Early Stopping
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        epochs_no_improve = 0
        torch.save(model.state_dict(), 'best_model.pth')
    else:
        epochs_no_improve += 1
        if epochs_no_improve >= patience:
            print(f"🛑 Early Stopping activado en la época {epoch+1}. La pérdida no mejoró en {patience} épocas.")
            break

print("\n✅ Entrenamiento finalizado.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix

# 1. Gráfico de Pérdida (Loss)
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history['train_loss'], label='Train Loss', color='#D96D55', lw=2)
plt.plot(history['val_loss'], label='Val Loss', color='#D99B77', lw=2)
plt.title('Evolución de la Pérdida (Loss)')
plt.xlabel('Épocas')
plt.ylabel('Loss')
plt.legend()
plt.grid(alpha=0.2)

# 2. Evaluación en Test
model.eval()
y_pred = []
with torch.no_grad():
    for inputs, _ in test_loader:
        outputs = model(inputs)
        _, predicted = torch.max(outputs, 1)
        y_pred.extend(predicted.numpy())

# 3. Matriz de Confusión
plt.subplot(1, 2, 2)
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Oranges', 
            xticklabels=le.classes_, yticklabels=le.classes_)
plt.title('Matriz de Confusión')
plt.xlabel('Predicción')
plt.ylabel('Real')
plt.tight_layout()
plt.show()

# 4. Reporte Detallado
print("\n--- REPORTE DE CLASIFICACIÓN ---")
print(classification_report(y_test, y_pred, target_names=le.classes_))
```
